# Problem 1

We have the function, $\mathrm{f\left(x_{1},x_{2}\right)=2x_{1}^{2}-4x_{1}x_{2}+1.5x_{2}^{2}+x_{2}}$

We first calculate the gradient and the Hessian,

$\mathrm{\nabla f\left(x_{1},x_{2}\right)=\begin{bmatrix}\frac{\partial f}{\partial x_{1}}\\
\frac{\partial f}{\partial x_{2}}
\end{bmatrix}=\begin{bmatrix}4x_{1}-4x_{2}+0+0\\
0-4x_{1}+3x_{2}+1
\end{bmatrix}=\begin{bmatrix}4x_{1}-4x_{2}\\
-4x_{1}+3x_{2}+1
\end{bmatrix}}$

$\mathrm{H\left(x_{1},x_{2}\right)=\begin{bmatrix}\frac{\partial^{2}f}{\partial x_{1}^{2}} & \frac{\partial^{2}f}{\partial x_{1}\partial x_{2}}\\
\frac{\partial^{2}f}{\partial x_{2}\partial x_{1}} & \frac{\partial^{2}f}{\partial x_{2}^{2}}
\end{bmatrix}=\begin{bmatrix}4-0 & 0-4\\
-4+0+0 & 0+3+0
\end{bmatrix}=\begin{bmatrix}4 & -4\\
-4 & 3
\end{bmatrix}}$

To find the stationary point, we set the gradient equal to zero and solve for $\mathrm{x_{1}}$ and $\mathrm{x_{2}}$,

$\mathrm{\nabla f\left(x_{1},x_{2}\right)=\begin{bmatrix}4x_{1}-4x_{2}\\
-4x_{1}+3x_{2}
\end{bmatrix}=\begin{bmatrix}0\\
-1
\end{bmatrix}}$  
From visual inspection, the stationary point is,  
$\mathrm{x_{1}=1}$ and $\mathrm{x_{2}=1}$

To show that this is a saddle point, we need to determine the eigenvalues of the Hessian.

$\mathrm{\left|H\left(x_{1},x_{2}\right)-\lambda I\right|=0}$

$\mathrm{\Rightarrow\begin{vmatrix}4-\lambda & -4\\
-4 & 3-\lambda
\end{vmatrix}=0}$

$\mathrm{\Rightarrow\left(4-\lambda\right)\left(3-\lambda\right)-16=0\Longrightarrow12-4\lambda-3\lambda+\lambda^{2}-16=0}$

$\mathrm{\Rightarrow\lambda^{2}-7\lambda-4=0}$

Using the quadratic formula we get,

$\mathrm{\lambda_{1}=\frac{7+\sqrt{7^{2}-4(1)(-4)}}{2(1)}=\frac{1}{2}\left(7+\sqrt{49+16}\right)=\frac{1}{2}\left(7+\sqrt{65}\right)\approx7.5311}$

$\mathrm{\lambda_{2}=\frac{7-\sqrt{7^{2}-4(1)(-4)}}{2(1)}=\frac{1}{2}\left(7-\sqrt{65}\right)\approx-0.5311}$

We have both a positive and negative eigenvalues, therefore this Hessian is an indefinite matrix. One direction curves up with the positive eigenvalue while the other curves down with the negative eigenvalue. Therefore, this is a saddle point.

Taylor-Series expansion:

$\mathrm{f\left(x_{1},x_{2}\right)\approx f\left(x_{0}\right)+\left.\nabla_{x}f\left(x\right)\right|_{x_{0}}^{T}\left(x-x_{0}\right)+\frac{1}{2}\left(x-x_{0}\right)^{T}\left.H\left(x_{1},x_{2}\right)\right|_{x_{0}}\left(x-x_{0}\right)}$

At $\mathrm{x_{1}=1}$ and $\mathrm{x_{2}=1}$

$\mathrm{f\left(x_{1},x_{2}\right)\approx f\left(1,1\right)+\begin{bmatrix}4x_{1}-4x_{2}\\
-4x_{1}+3x_{2}+1
\end{bmatrix}^{T}\begin{bmatrix}x_{1}-1\\
x_{2}-1
\end{bmatrix}+\frac{1}{2}\begin{bmatrix}x_{1}-1\\
x_{2}-1
\end{bmatrix}^{T}\begin{bmatrix}4 & -4\\

x_{2}-1
\end{bmatrix}}$

$\mathrm{\Rightarrow f\left(x_{1},x_{2}\right)\approx f\left(1,1\right)+\begin{bmatrix}4\left(1\right)-4\left(1\right)\\
-4\left(1\right)+3\left(1\right)+1
\end{bmatrix}^{T}\begin{bmatrix}x_{1}-1\\
x_{2}-1
\end{bmatrix}+\frac{1}{2}\begin{bmatrix}x_{1}-1\\
x_{2}-1
\end{bmatrix}^{T}\begin{bmatrix}4 & -4\\
-4 & 3
\end{bmatrix}\begin{bmatrix}x_{1}-1\\

\end{bmatrix}}$

$\mathrm{\Rightarrow f\left(x_{1},x_{2}\right)\approx f\left(1,1\right)+\begin{bmatrix}0\\
0
\end{bmatrix}^{T}\begin{bmatrix}x_{1}-1\\

\end{bmatrix}^{T}\begin{bmatrix}4 & -4\\
-4 & 3
\end{bmatrix}\begin{bmatrix}x_{1}-1\\
x_{2}-1

$\mathrm{\Rightarrow f\left(x_{1},x_{2}\right)\approx f\left(1,1\right)+\frac{1}{2}\begin{bmatrix}x_{1}-1 & x_{2}-1\end{bmatrix}\begin{bmatrix}4 & -4\\
-4 & 3
\end{bmatrix}\begin{bmatrix}x_{1}-1\\

\end{bmatrix}}$

$\mathrm{\Rightarrow f\left(x_{1},x_{2}\right)\approx f\left(1,1\right)+\frac{1}{2}\begin{bmatrix}4x_{1}-4x_{2} & -4x_{1}+3x_{2}+1\end{bmatrix}\begin{bmatrix}x_{1}-1\\
x_{2}-1
\end{bmatrix}}$

$\mathrm{\Rightarrow f\left(x_{1},x_{2}\right)\approx f\left(1,1\right)+\frac{1}{2}\left[\left(4x_{1}-4x_{2}\right)\left(x_{1}-1\right)+\left(-4x_{1}+3x_{2}+1\right)\left(x_{2}-1\right)\right]}$

Expanding this out, we get,

$\mathrm{\Rightarrow f\left(x_{1},x_{2}\right)-f\left(1,1\right)\approx \frac{1}{2}\left[4x_{1}^{2}-4x_{1}-4x_{1}x_{2}+4x_{2}-4x_{1}x_{2}+4x_{1}+3x_{2}^{2}-3x_{2}+x_{2}-1\right]}$

$\mathrm{\Rightarrow2x_{1}^{2}+1.5x_{2}^{2}-4x_{1}x_{2}+x_{2}-0.5}$

We are looking for $\mathrm{f\left(x_{1},x_{2}\right)<0}$, so we can start by setting our function equal to zero, giving,

$\mathrm{f\left(x_{1},x_{2}\right)-f\left(1,1\right)\approx2x_{1}^{2}+1.5x_{2}^{2}-4x_{1}x_{2}+x_{2}-0.5=0}$

We can divide by $\mathrm{x_{2}^{2}}$, which gives,

$\mathrm{\Rightarrow2\frac{x_{1}^{2}}{x_{2}^{2}}+1.5-4\frac{x_{1}}{x_{2}}+1-0.5=0}$

Let $\mathrm{u\equiv\frac{x_{1}}{x_{2}}}$, then,

$\mathrm{\Rightarrow2u^{2}-4u+2=0}$

If we define $\mathrm{u_{1}}$ and $\mathrm{u_{2}}$ as the roots to this quadratic function, then the L.H.S. can be expressed as,

$\mathrm{\left(u-u_{1}\right)\left(u-u_{2}\right)=0}$

Thus,

$\mathrm{f\left(x_{1},x_{2}\right)-f\left(1,1\right)\approx\left(u-u_{1}\right)\left(u-u_{2}\right)=0}$

Expanding out and substituting back in for u, gives,

$\mathrm{u^{2}-uu_{1}-uu_{2}+u_{1}u_{2}=0}$

$\mathrm{\Rightarrow\frac{x_{1}^{2}}{x_{2}^{2}}-u_{1}\frac{x_{1}}{x_{2}}-u_{2}\frac{x_{1}}{x_{2}}+u_{1}u_{2}=0}$

Multiplying by $\mathrm{x_{2}^{2}}$,

$\mathrm{x_{1}^{2}-u_{1}x_{1}x_{2}-u_{2}x_{1}x_{2}+u_{1}u_{2}x_{2}^{2}=0}$

$\mathrm{\Rightarrow x_{1}^{2}-\left(u_{1}+u_{2}\right)x_{1}x_{2}+u_{1}u_{2}x_{2}^{2}=0}$

Factoring this again gives,

$\mathrm{\left(x_{1}-u_{1}x_{2}\right)\left(x_{1}-u_{2}x_{2}\right)=0}$

Thus,

$\mathrm{f\left(x_{1},x_{2}\right)-f\left(1,1\right)\approx\left(x_{1}-u_{1}x_{2}\right)\left(x_{1}-u_{2}x_{2}\right)<0}$
