We have the function, $\mathrm{f\left(x_{1},x_{2}\right)=2x_{1}^{2}-4x_{1}x_{2}+1.5x_{2}^{2}+x_{2}}$

We first calculate the gradient and the Hessian,

$\mathrm{\nabla f\left(x_{1},x_{2}\right)=\begin{bmatrix}\frac{\partial f}{\partial x_{1}}\\
\frac{\partial f}{\partial x_{2}}
\end{bmatrix}=\begin{bmatrix}4x_{1}-4x_{2}+0+0\\
0-4x_{1}+3x_{2}+1
\end{bmatrix}=\begin{bmatrix}4x_{1}-4x_{2}\\
-4x_{1}+3x_{2}+1
\end{bmatrix}}$

$\mathrm{H\left(x_{1},x_{2}\right)=\begin{bmatrix}\frac{\partial^{2}f}{\partial x_{1}^{2}} & \frac{\partial^{2}f}{\partial x_{1}\partial x_{2}}\\
\frac{\partial^{2}f}{\partial x_{2}\partial x_{1}} & \frac{\partial^{2}f}{\partial x_{2}^{2}}
\end{bmatrix}=\begin{bmatrix}4-0 & 0-4\\
-4+0+0 & 0+3+0
\end{bmatrix}=\begin{bmatrix}4 & -4\\
-4 & 3
\end{bmatrix}}$

To find the stationary point, we set the gradient equal to zero and solve for $\mathrm{x_{1}}$ and $\mathrm{x_{2}}$,

$\mathrm{\nabla f\left(x_{1},x_{2}\right)=\begin{bmatrix}4x_{1}-4x_{2}\\
-4x_{1}+3x_{2}
\end{bmatrix}=\begin{bmatrix}0\\
-1
\end{bmatrix}}$  
From visual inspection, the stationary point is,  
$\mathrm{x_{1}=1}$ and $\mathrm{x_{2}=1}$

To show that this is a saddle point, we need to determine the eigenvalues of the Hessian.

$\mathrm{\left|H\left(x_{1},x_{2}\right)-\lambda I\right|=0}$

$\mathrm{\Rightarrow\begin{vmatrix}4-\lambda & -4\\
-4 & 3-\lambda
\end{vmatrix}=0}$

$\mathrm{\Rightarrow\left(4-\lambda\right)\left(3-\lambda\right)-16=0\Longrightarrow12-4\lambda-3\lambda+\lambda^{2}-16=0}$

$\mathrm{\Rightarrow\lambda^{2}-7\lambda-4=0}$

Using the quadratic formula we get,

$\mathrm{\lambda_{1}=\frac{7+\sqrt{7^{2}-4(1)(-4)}}{2(1)}=\frac{1}{2}\left(7+\sqrt{49+16}\right)=\frac{1}{2}\left(7+\sqrt{65}\right)\approx7.5311}$

$\mathrm{\lambda_{2}=\frac{7-\sqrt{7^{2}-4(1)(-4)}}{2(1)}=\frac{1}{2}\left(7-\sqrt{65}\right)\approx-0.5311}$

We have both a positive and negative eigenvalues, therefore this Hessian is an indefinite matrix. One direction curves up with the positive eigenvalue while the other curves down with the negative eigenvalue. Therefore, this is a saddle point.

Taylor-Series expansion:

### a)

Given the plane $\mathrm{x_{1}+2x_{2}+3x_{3}=1}$, we can setup an unconstrained optimzation problem as the following.

$\mathrm{min\;\;\left\Vert x-x_{0}\right\Vert ^{2}}$

Where,

$\mathrm{x_{0}}=\begin{bmatrix}-1\\
0\\
1
\end{bmatrix}$

To turn this into an unconstrained optimziation problem, we can express our optimization function f(x) as a function of $\mathrm{x_{2}}$ and $\mathrm{x_{3}}$ only. This is done by rearranging the equation of the plane to solve for $\mathrm{x_{1}}$

$\mathrm{x_{1}=-2x_{2}-3x_{3}+1}$

Substituting this into our optimization function, we have,

$\mathrm{f\left(x\right)=\left\Vert x-x_{0}\right\Vert ^{2}=\left(x_{1}-x_{0}\right)^{2}+\left(x_{2}-x_{0}\right)^{2}+\left(x_{3}-x_{0}\right)^{2}}$  
$\mathrm{\Rightarrow\left(-2x_{2}-3x_{3}+1-x_{0}\right)^{2}+\left(x_{2}-x_{0}\right)^{2}+\left(x_{3}-x_{0}\right)^{2}}$

Then we can substitute in the values of $\mathrm{x_{0}}$,

$\mathrm{f\left(x\right)=\left(-2x_{2}-3x_{3}+2\right)^{2}+x_{2}^{2}+\left(x_{3}-1\right)^{2}}$

Expanding out and combining terms, our optimization function becomes,

$\mathrm{f\left(x\right)=\left(-2x_{2}-3x_{3}+2\right)\left(-2x_{2}-3x_{3}+2\right)+x_{2}^{2}+\left(x_{3}-1\right)\left(x_{3}-1\right)}$

$\mathrm{\Rightarrow\;=4x_{2}^{2}+12x_{2}x_{3}-8x_{2}+9x_{3}^{2}-12x_{3}+4+x_{2}^{2}+x_{3}^{2}-2x_{3}+1}$

$\mathrm{\Rightarrow\;f\left(x\right)=5x_{2}^{2}+10x_{3}^{2}+12x_{2}x_{3}-8x_{2}-14x_{3}+5}$

**Thus our unconstrained optimization problem becomes,**

$\mathrm{min:\;\;f\left(x\right)}$

Where,

$\mathrm{f\left(x\right)=5x_{2}^{2}+10x_{3}^{2}+12x_{2}x_{3}-8x_{2}-14x_{3}+5}$

$\mathrm{x_{1}=-2x_{2}-3x_{3}+1}$

This problem is strictly convex. This can be found by looking at the Hessian. First, we find the gradient. Then we calculate the Hessian,

$\mathrm{\nabla_{x}f\left(x\right)=\begin{bmatrix}\frac{\partial f}{\partial x_{2}}\\
\frac{\partial f}{\partial x_{3}}
\end{bmatrix}=\begin{bmatrix}10x_{2}+12x_{3}-8\\
20x_{3}+12x_{2}-14
\end{bmatrix}}$

$\mathrm{H\left(x\right)=\begin{bmatrix}\frac{\partial^{2}f}{\partial x_{2}^{2}} & \frac{\partial^{2}f}{\partial x_{2}\partial x_{3}}\\
\frac{\partial^{2}f}{\partial x_{3}\partial x_{2}} & \frac{\partial^{2}f}{\partial x_{3}^{2}}
\end{bmatrix}=\begin{bmatrix}10 & 12\\
12 & 20
\end{bmatrix}}$

Next, we need to determine the eigenvalues of the Hessian. Determining the characteristic polynomial and then solving it for eigenvalues, gives,

$\mathrm{\left|H-\lambda I\right|=0\Longrightarrow\begin{vmatrix}10-\lambda & 12\\
12 & 20-\lambda
\end{vmatrix}=\left(10-\lambda\right)\left(20-\lambda\right)-\left(12\right)\left(12\right)=200-30\lambda+\lambda^{2}-144=\lambda^{2}-30\lambda+56}$

$\mathrm{\lambda^{2}-30\lambda+56=0\Longrightarrow\left(\lambda-28\right)\left(\lambda-2\right)=0}$

$\mathrm{\lambda_{1}=28}$ and $\mathrm{\lambda_{2}=2}$

Since all of our eigenvalues are greater than zero, this Hessian is p.d. Further, because this Hessian is constant, it will be p.d. everywhere, thus making our optimization function strictly convex. Also, since there are no constraints, we do not need to check if they are convex. Therefore, this optimzation problem is strictly convex.

```python
# MAE 598 - Design Optimization HW # 2
# Benjamin Webb
# 9/16/2022

import numpy as np
import matplotlib.pyplot as plt


def gradx(x):
	# Calculates gradient for problem 2
	# given input vector x and x0
	# x = current point

	g = np.zeros((2, 1))
	g[0] = 10*x[0] + 12*x[1] - 8.0
	g[1] = 20*x[1] + 12*x[0] - 14.0

	return g


def hessian():
	# Generates Hessian for problem 2
	# Hessian is constant for this problem

	H = np.array([[10.0, 12.0], [12.0, 20.0]])

	return H


def GD_inexact(t, alpha, x0, epsilon0):
	# Performs gradient desecent using inexact line search for problem 2
	# t: tuning parameter between (0, 1)
	# alpha: initial step-size
	# x0: initial starting point
	# epsilon0: stopping criteria

	# Initialize stopping critera
	epsilon = np.zeros((1000, 1))
	epsilon[0] = 1.0
	n = np.uint16(0)

	# Begin gradient descent loop
	while epsilon[n] > epsilon0:
		# Problem 2 minimize function ||x - x0||^2
		f0 = 5.0*x0[0]**2 + 10.0*x0[1]**2 + 12.0*x0[0]*x0[1] - 8.0*x0[0] - 14.0*x0[1] + 5.0

		# Calculate gradient at current point
		g0 = gradx(x0)

		# Calculate f(x - alpha*g0)
		x = x0 - alpha*g0
		f = 5.0*x[0]**2 + 10.0*x[1]**2 + 12.0*x[0]*x[1] - 8.0*x[0] - 14.0*x[1] + 5.0

		# Calculate Phi(alpha) = f0 - t*g0^T*g0*alpha
		phi = f0 - t * g0.T @ g0 * alpha

		# Determine how to proceed to next step
		if np.abs(f) > np.abs(phi):
			# Update alpha
			alpha = 0.5 * alpha
		elif n >= 1000:
			break
		else:
			n = n + np.uint16(1)
			epsilon[n] = np.linalg.norm(f - f0)
			x0 = x
			alpha = 1.0

	# Print complete statement with # of iterations
	print('Gradient Descent Optimization Complete')
	print('# of Iterations: ' + str(n))
	return x, n, epsilon[1:n+1]

def newton(x0, epsilon0):
	# Function performs Newton's method for optimization of problem 2
	# x0: initial guess
	# epsilon0: stopping criteria

	# Initialize stopping criteria
	epsilon = np.zeros((1000, 1))
	epsilon[0] = 1.0
	n = np.uint16(0)

	# Calculate Hessian once, since it is constant over domain of this function
	H0 = hessian()
	H0_1 = np.linalg.inv(H0)

	# Begin optimization loop
	while epsilon[n] > epsilon0:
		# Problem 2 optimzation function
		f0 = 5.0*x0[0]**2 + 10.0*x0[1]**2 + 12.0*x0[0]*x0[1] - 8.0*x0[0] - 14.0*x0[1] + 5.0

		# Calculate gradient at current point
		g0 = gradx(x0)

		# Calculate next x values
		x = x0 - H0_1 @ g0

		# Update optimization function value
		f = f0 - 0.5*g0.T @ H0_1 @ g0

		# Update stoping criteria and determine next step
		n = n + np.uint16(1)
		epsilon[n] = np.linalg.norm(f - f0)
		x0 = x
		if n >= 1000:
			break

	# Print complete statement with # of iterations
	print('Newton Optimization Complete')
	print('# of Iterations: ' + str(n))
	return x, n, epsilon[1:n+1]


# Main program for problem 2
if __name__ == '__main__':
	# HW2 Problem 2

	# Gradient descent w/ inexact line search with t = 1.0
	x_GD, n_GD, eps_GD = GD_inexact(1.0, 1.0, np.array([[10.0], [10.0]]), 0.000001)
	x1_GD = -2.0*x_GD[0] - 3.0*x_GD[1] + 1

	print('t = ' + str(1.0))
	print('x1 = ' + str(x1_GD))
	print('x2 = ' + str(x_GD[0]))
	print('x3 = ' + str(x_GD[1]))

	# Newton's method
	x_N, n_N, eps_N = newton(np.array([[10.0], [10.0]]), 0.000001)
	x1_N = -2.0*x_N[0] - 3.0*x_N[1] + 1

	print('x1 = ' + str(x1_N))
	print('x2 = ' + str(x_N[0]))
	print('x3 = ' + str(x_N[1]))

	# Compare solutions of each method
	# Determine distance of each solution to desired point (-1, 0, 1)^T
	d_GD = np.sqrt((x1_GD + 1.0)**2 + x_GD[0]**2 + (x_GD[1] - 1.0)**2)
	d_N = np.sqrt((x1_N + 1.0)**2 + x_N[0]**2 + (x_N[1] - 1.0)**2)
	print(' ')
	print('Gradient Descent Distance = ' + str(d_GD))
	print('Newtons Method Distance = ' + str(d_N))
	print(' ')

	# Plot method convergence
	fig = plt.figure(num=1)
	plt.semilogy(np.linspace(1, n_GD, n_GD), eps_GD, 'bD--', np.linspace(1, n_N, n_N), eps_N, 'rD--')
	plt.grid(b=True, which='both', axis='y')
	plt.title('t = 1.0')
	plt.ylabel('error')
	plt.xlabel('# of Iterations')
	fig.legend(['Gradient Descent', 'Newtons Method'])

	# Gradient descent w/ inexact line search with t = 0.5
	x_GD, n_GD, eps_GD = GD_inexact(0.5, 1.0, np.array([[10.0], [10.0]]), 0.000001)
	x1_GD = -2.0 * x_GD[0] - 3.0 * x_GD[1] + 1

	print('t = ' + str(0.5))
	print('x1 = ' + str(x1_GD))
	print('x2 = ' + str(x_GD[0]))
	print('x3 = ' + str(x_GD[1]))

	# Determine distance of each solution to desired point (-1, 0, 1)^T
	d_GD = np.sqrt((x1_GD + 1.0) ** 2 + x_GD[0] ** 2 + (x_GD[1] - 1.0) ** 2)
	print(' ')
	print('Gradient Descent Distance = ' + str(d_GD))
	print(' ')

	# Plot method convergence
	fig = plt.figure(num=2)
	plt.semilogy(np.linspace(1, n_GD, n_GD), eps_GD, 'bD--', np.linspace(1, n_N, n_N), eps_N, 'rD--')
	plt.grid(b=True, which='both', axis='y')
	plt.title('t = 0.5')
	plt.ylabel('error')
	plt.xlabel('# of Iterations')
	fig.legend(['Gradient Descent', 'Newtons Method'])

	# Gradient descent w/ inexact line search with t = 0.3
	x_GD, n_GD, eps_GD = GD_inexact(0.3, 1.0, np.array([[10.0], [10.0]]), 0.000001)
	x1_GD = -2.0 * x_GD[0] - 3.0 * x_GD[1] + 1

	print('t = ' + str(0.3))
	print('x1 = ' + str(x1_GD))
	print('x2 = ' + str(x_GD[0]))
	print('x3 = ' + str(x_GD[1]))

	# Determine distance of each solution to desired point (-1, 0, 1)^T
	d_GD = np.sqrt((x1_GD + 1.0) ** 2 + x_GD[0] ** 2 + (x_GD[1] - 1.0) ** 2)
	print(' ')
	print('Gradient Descent Distance = ' + str(d_GD))
	print(' ')

	# Plot method convergence
	fig = plt.figure(num=3)
	plt.semilogy(np.linspace(1, n_GD, n_GD), eps_GD, 'bD--', np.linspace(1, n_N, n_N), eps_N, 'rD--')
	plt.grid(b=True, which='both', axis='y')
	plt.title('t = 0.3')
	plt.ylabel('error')
	plt.xlabel('# of Iterations')
	fig.legend(['Gradient Descent', 'Newtons Method'])
```

Given the function, $\mathrm{h\left(I,I_{t}\right)=\left\{ \begin{array}{cc}

### a)

Given h, we can plot h for $\mathrm{0\leq I\leq10}$, as an example. We know from the context of the problem that I is light intensity and $\mathrm{I_{t}}$ is the target light intensity. As such, we know that I and $\mathrm{I_{t}}$ will never be negative.

```python
# Plot for problem 4

import matplotlib.pyplot as plt
import numpy as np

if __name__ == '__main__':
	# Let's plot

	# Make some vectors for I and It
	I = np.linspace(0.0, 10.0, 101)
	I[0] = 0.001
	It = np.ones((101, 1))*5.0

	# Evaluate h(I, It)
	h = np.zeros((101, 1))
	for j in range(0, 101, 1):
		if I[j] <= It[j]:
			h[j] = It[j] / I[j]
		else:
			h[j] = I[j] / It[j]

	fig = plt.figure(num=1)
	plt.plot(I[0:50], h[0:50], '-b', I[50:100], h[50:100], '-r')
	plt.grid(b=True, which='both', axis='both')
	plt.xlabel('I')
	plt.ylabel('h(I,It)')
	plt.xlim((0.0, 10.0))
	plt.ylim((0.0, 2.5))
	plt.legend(['I <= It', 'I >= It'])
	plt.show()
```

As we can see from the plot, h(I,It) is convex and has a minimum value of 1. This minimum value occurs when I = $\mathrm{I_{t}}$. So we know that h is convex w.r.t. I and $\mathrm{I_{t}}$. To determine if our optimization problem is convex, we need to determine if h is convex w.r.t. $\mathrm{a^{T}p}$. More specifically, we need to determine if h is convex w.r.t. to p. To do this, we need to look at the gradient and the Hessian of h.


$\mathrm{\nabla_{p}h\left(a^{T}p,I_{t}\right)=\frac{dh}{dI}\frac{\partial\left(a^{T}p\right)}{\partial p}=\nabla_{I}h\cdot a}$

$\mathrm{H\left(h\right)=\frac{\partial^{2}h}{\partial p^{2}}=\frac{\partial}{\partial p}\left(\nabla_{I}h\cdot a\right)=\frac{d\left(\nabla_{I}h\right)}{I}\frac{\partial\left(a^{T}p\right)}{\partial p}\cdot a^{T}=\nabla_{I}^{2}h\,a\,a^{T}}$

As we showed in the plot above, h is convex, thus $\mathrm{\nabla_{I}^{2}h\geq0}$, making it p.s.d. and it is also a scalar. We showed in class that the matrix, $\mathrm{a\,a^{T}}$, is also p.s.d. Therefore, h w.r.t. p is convex.

In class, we also showed that, if $\mathrm{f_{1},f_{2}}$ are convex then, $\mathrm{max\begin{Bmatrix}f_{1} & f_{2}\end{Bmatrix}}$ is convex. Therefore, our optimization problem **is convex**, since our only constraint is also convex.

We saw from the plot, that the minimum value of h occurs when $\mathrm{I=I_{t}}$.

Thus, when $\mathrm{I\leq I_{t}}$, h is always decreasing until $\mathrm{I=I_{t}}$. When $\mathrm{I\geq I_{t}}$, the negative gradient of h will cause the optimization to always slope down towards $\mathrm{I=I_{t}}$. This combination suggests that this problem **will have a unique solution**, specifically a solution that gives $\mathrm{I=I_{t}}$.
